```python
lineal.plot.scatter(x="metro", y="precio")
plt.show
```

```python
regresion = linear_model.LinearRegression()
```

```python
precios = lineal["precio"].values.reshape((-1,1))
```

```python
model = regresion.fit(precios, lineal["metro"])
```

```python
print(model.intercept_)
print(model.coef_)
```

```python
data = [[300],[487],[600]]
predi = model.predict(data)
print(predi)
```

```python
lineal.plot.scatter(x="precio", y="metro")
plt.plot(data, predi, color='black')
plt.show
```
